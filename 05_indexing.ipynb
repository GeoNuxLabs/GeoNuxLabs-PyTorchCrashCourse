# Tensor indexing

As you now know, a tensor is basically a matrix that is fully compatible with PyTorch operations. If you come from a GIS background, or any field that works with matrices, you already know that indexing is essential for being able to extract, manipulate, and understand data.

This lecture will walk through basic indexing, which you will recognize from NumPy, and then move on to more PyTorch‑specific indexing functions. 

**See the code below, and refresh your memory on basic indexing.** 

```python
import torch

# A simple 2×3 tensor
grades = torch.tensor([[20, 25, 30],
                       [30, 20, 25]])

# Get value at row 0, column 1
value = grades[0, 1]
print("Value at [0, 1]:", value)

# Entire row (row 1)
row = grades[1]
print("Row 1:", row)

# Entire column (column 2)
column = grades[:, 2]
print("Column 2:", column)

```

## Finding the index of the highest value

When working with tensors, it’s often not enough to know what the highest value is, you also need to know where it is. This is crucial in deep learning, because the final prediction of a model is usually the index of the largest value in the output tensor.

PyTorch makes this extremely simple with the argmax() function.
Combined with the dim argument, it becomes one of the most powerful indexing tools you will use.

**How to think**
Argmax() search for the index of the largest value 
dim=1 search horizontally, row by row

```python
# Index 2 on row 0, and index 0 on row 1 is highest
import torch

grades = torch.tensor([[20, 25, 30],
                       [30, 20, 25]])

# Find the index of the highest grade in each row
best_grades = torch.argmax(grades, dim=1)

print(best_grades)
```

## Expert level indexing

In some cases, you may want to be very specific in your indexing and pick different elements from each row. For example:

- From row 0, get the value at column 2
- From row 1, get the value at column 0
- From row 2, get the value at column 3

You could do this with a simple for‑loop, but that would be inefficient when working with large tensors inside a deep learning pipeline. Instead, PyTorch provides a convenient and highly optimized function for this purpose: torch.gather().

This function allows you to collect different elements from each row (or column) in a single, vectorized operation, no loops needed.

```python
grades = torch.tensor([[20, 25, 30],
                       [30, 20, 25],
                       [15, 10, 40]])

# The specified indices we want to fetch values from 
# from each row
selected_indices = torch.tensor([[2], [0], [1]])

# Gather from grades along columns (dim=1)
selected_values = torch.gather(grades, dim=1,
                                index=selected_indices)

print(selected_values)
```

The code uses torch.gather() to pick different elements from each row of the tensor. The grades tensor has three rows, and the selected_indices tensor tells PyTorch which column to pick from each row. Because dim=1 is used, the gathering happens horizontally across the columns. This means: from row 0, take the value at column 2; from row 1, take the value at column 0; and from row 2, take the value at column 1. The result is a new tensor containing exactly those three selected values.

**Note**: 
Understanding torch.gather() is very important. When developing more complex models, this operation becomes essential because it lets you extract different elements from each row (or column) in a single, efficient, vectorized step. Without it, you would be forced to use slow Python loops, which quickly become a bottleneck in deep learning workflows.
