# Deep Learning for GeoAI in PyTorch

#### DL in PyTorch is based on five distinct steps, which all have a math concept calculated on raw tensors. These steps can all be automated and proffesionaly handled in PyTorch with "torch.nn".  

## Step 1 — Prediction

A prediction is computed at the neuron level.  
Each neuron receives an input and produces an output based on an activation function.

The basic prediction formula is:


$\hat{y} = f(X, W)$


Where:

- **X** — the input features  
- **W** — the weights associated with the neuron  
- **f** — the activation function  
- $\hat{y}$ — the predicted output  


#### Raw tensor calculation
From scratch, the raw prediction (before applying any activation function) is computed using tensors as:



y_hat = X @ W + b




Where:

- **X** the input features
- **@** denotes matrix multiplication
- **W** the weights associated with the neuron   
- **b** is the bias term  
- **y_hat** is the raw, unactivated prediction  


#### Using the torch.nn module
PyTorch provides an easy and professional alternative to raw tensor calculations through the `torch.nn` module.  
Instead of manually computing predictions using matrix multiplication, PyTorch allows you to define a model and call it directly:

y_hat = model(x)


This automatically performs:

- the forward pass  
- all internal linear layers  
- activation functions  
- bias terms  
- and any other modules you have defined  

This makes the code cleaner, more modular, and easier to maintain compared to manual tensor math.

---


## Step 2 — Loss Calculation

Loss is a continuous metric that quantifies how far a prediction is from the true value.  
Even if the prediction task is binary, the loss function itself is always continuous.

The conceptual definition of loss is:

$L = Loss(\hat{y}, y)$

Where:

- **L** — the quantified error  
- **$\hat{y}$** — the predicted output  
- **y** — the ground truth  
- **Loss** — a function that measures the difference between prediction and truth. Common examples include:

    - **MSE (Mean Squared Error)** for regression  
    - **Cross‑Entropy Loss** for classification
    - **Binary Cross‑Entropy Loss** for binary          classification   


#### Raw tensor loss calculation
In raw tensor form, the loss can be computed manually as:

loss = torch.mean((y_hat - y)**2)

Where:

- **$\hat{y}$** — the predicted output  
- **y** — the ground truth 

This corresponds to Mean Squared Error (MSE)


#### Using the torch.nn module
PyTorch provides built‑in loss functions through the `torch.nn` module. After defining a loss function (criterion), the loss is computed as:

loss = criterion(y_hat,y)

Where: 

- **criterion** — pre-defined loss function
- **$\hat{y}$** — the predicted output  
- **y** — the ground truth 

This produces the same conceptual result but is cleaner, safer, and integrates with the rest of the PyTorch training pipeline.

---

## Step 3 — Gradient Calculation

In order for the model to understand how wrong the prediction is and how to adjust its weights and biases, we need to compute the gradient of the loss function.  
The gradient tells us the slope of the loss — in other words, the direction and magnitude of change needed to reduce the error.

Conceptually, the gradient is written as:

∇L  
(pronounced “nabla L” or “the gradient of the loss”)

Where:

- ∇ — the gradient operator  
- L — the loss value  
- The result is a set of partial derivatives for every weight and bias in the model  

This gradient is what allows backpropagation to update parameters in the correct direction.


#### Raw tensor gradient calculation
In raw tensor form, PyTorch computes gradients automatically.  
Once the loss is calculated, you trigger gradient computation with:

loss.backward()

This performs:

- the full backward pass  
- gradient calculation for every parameter  
- accumulation of gradients in param.grad  

You can inspect gradients manually:

print(model.linear.weight.grad)


#### Using the torch.nn module
PyTorch’s torch.nn module computes gradients in the same conceptual way as raw tensors —  
the difference is that it handles all the bookkeeping for you.

After defining a model and a loss function, the gradient step is still:

loss.backward()

This integrates seamlessly with optimizers such as SGD or Adam, which use the gradients to update the model parameters.

---

## Step 4 — Parameter Update

Once the gradients have been calculated in the backward pass, the model needs to update its parameters (weights and biases).  
This step is based on the mathematical idea of **stepping down the slope** of the loss function.

The general update rule is:

θₜ₊₁ = θₜ − η∇0L

Where:

- θ — a parameter (weight or bias)  
- η — the learning rate  
- ∇θL — the gradient of the loss with respect to that parameter  
- θₜ₊₁ — the updated parameter value  

This operation moves each parameter **in the direction that reduces the loss**.


#### Parameter update with raw tensors
Using raw tensors, a weight update looks like this:

W -= lr * W.grad

This means:

- take the current weight  
- subtract the gradient scaled by the learning rate  
- store the updated weight back into W  

This is the core of gradient descent.


#### Parameter update with torch.nn and optimizers
When using PyTorch’s optimizer system, the update step is handled automatically.  
After computing gradients with loss.backward(), you update all parameters with:

optimizer.step()

This applies the same mathematical rule as above, but in a cleaner and more modular way.

Common optimizers include:

- **SGD** — simple gradient descent with optional momentum  
- **Adam** — adaptive learning rate optimizer  

These optimizers use the gradients to adjust the model parameters in an efficient and stable way.

---


## Step 5 — Gradient Reset

Before starting the next training loop (epoch), the gradients must be reset.  
This step has no mathematical concept behind it, it is simply a technical requirement.  
If we do not reset the gradients, PyTorch will accumulate them from previous iterations, which leads to incorrect updates.


#### With raw tensors

w.grad.zero_()

This clears the gradient stored in the tensor so that the next backward pass starts from zero.


#### With torch.nn

optimizer.zero_grad()

This resets the gradients for all parameters managed by the optimizer.  
It prepares the model for the next forward–backward–update cycle.


#### Why this step is necessary

PyTorch accumulates gradients by default.  
This means:

new_gradients = old_gradients + current_gradients

Resetting ensures that each training step uses only the gradients from the current batch.  
Without this reset, training would diverge instead of converging.

---

## Summary — The Full Training Cycle

A full training iteration (one batch or one epoch) consists of the same five steps repeated in order.  
Each step plays a specific role in how a neural network learns.

1. **Prediction**  
   The model takes input data and produces an output (prediction).  
   This is the forward pass where the network computes y_hat.

2. **Loss Calculation**  
   The loss function measures how far the prediction is from the true value.  
   This gives a continuous error signal L.

3. **Gradient Calculation**  
   Backpropagation computes the gradients of the loss with respect to every weight and bias.  
   These gradients tell the model how to change its parameters.

4. **Parameter Update**  
   The optimizer updates the parameters using the gradients.  
   Raw tensors use: W -= lr * W.grad  
   torch.nn uses: optimizer.step()

5. **Gradient Reset**  
   Gradients are cleared before the next iteration.  
   Raw tensors use: w.grad.zero_()  
   torch.nn uses: optimizer.zero_grad()

These five steps form the core of every training loop in PyTorch.  
They repeat for every batch and every epoch until the model converges.
