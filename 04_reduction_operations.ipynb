# Tensor reduction operations

A reduction is any operation that reduces a tensor to a smaller number of elements. Common examples include:

- sum()
- mean()
- max()

In this lecture, we will explore how the dim argument influences reduction operations.
In short, the dim argument determines along which axis the tensor collapses when a reduction is applied.

Before diving into that, let’s first understand how reduction operations affect a tensor when no dim argument is provided.

```python
import torch

# A 2×3 tensor representing grades for two persons
# across three tasks
grades = torch.tensor([[20., 25., 30.],
                       [30., 20., 25.]])

# Reduction operations WITHOUT specifying dim:
# These operate on ALL elements in the tensor.

mean_value = grades.mean() # Mean of all numbers
sum_value = grades.sum() # Sum of all numbers
max_value = grades.max() # Highest value in the tensor

print(f"Overall mean: {mean_value}")
print(f"Sum of all values in tensor: {sum_value}")
print(f"Highest value in tensor: {max_value}")
```

 ---
## The **dim** argument for **2D tensors**

When PyTorch performs reduction operations, you can control which direction the tensor collapses. In other words, you decide whether PyTorch should combine values row‑by‑row or column‑by‑column

- no dim argument
    - The entire tensor collapses into a single value.

- dim = 0
    - Collapses the rows, operating vertically. The result is a vector whose length equals the number of columns.

- dim = 1
    - Collapses the columns, operating horizontally. The result is a vector whose length equals the number of rows.
