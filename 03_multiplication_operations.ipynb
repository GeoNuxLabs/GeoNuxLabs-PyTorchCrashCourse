# Tensor multiplication operations

From the previous lectures we have learned that the tensor is the fundamental data structure of all deep learning. A tensor can represent both learnable parameters and regular data, depending on whether we set requires_grad=True. We have also seen that whenever a tensor with requires_grad=True is used in a computation, PyTorch records that operation in a **computational graph**. This graph is later used to compute the loss and the gradients during backpropagation (Autograd).

To make use of these PyTorch mechanisms, we need to understand how to perform meaningful operations on tensors. Fortunately, deep learning relies heavily on a small set of operations that repeat over and over again. Once you understand these, you understand most of the math behind neural networks.

We will start by distinguishing between two fundamental types of tensor operations:

- **Element-wise multiplication (*)**  
  Multiplies each element in one tensor with the corresponding element in another tensor.  
  Example:  
    - [a, b] * [c, d] = [ac, bd]


- **Matrix multiplication (@)**  
Performs true matrix multiplication, combining rows and columns according to linear algebra rules.  
This is the operation used in neural network layers.  
Example:  
    - [[a, b], [[e], [[ae + bg], [c, d]] @ [f]] = [ce+dg]][g]]


Understanding the difference between these two operations is essential, because neural networks rely almost entirely on matrix multiplication to transform inputs, while element-wise operations are used for activations, scaling, and other transformations.


 ---
## Element-wise multiplication (*)

Element‑wise multiplication is the simplest type of tensor multiplication.  
It takes **two tensors of the same shape** and multiplies each element with the element in the corresponding position.

In other words, the tensors are aligned on top of each other, and PyTorch multiplies the overlapping values one by one.

**Important:** The tensors **must** have the same shape for element‑wise multiplication (unless broadcasting is used, which we will cover later).

This operation is extremely common in deep learning, especially in:

- activation functions

- scaling operations

- attention mechanisms

- normalization layers

It is one of the core building blocks of tensor computation.

```python
import torch 

a = torch.tensor([[1,2], [3,4]])
b = torch.tensor([[10, 20], [30,40]])

element_multiplied = a*b 

# Note that overlapping values will be multiplied 
print(element_multiplied)
```

 --- 
## Matrix multiplication (@)

Matrix multiplication is one of the most important operations in deep learning.  
Every linear layer in a neural network is powered by this operator, so understanding it is essential.

The key rule to remember is:

**The number of columns in the first matrix must equal the number of rows in the second matrix.**

This is the “red thread” of matrix multiplication:

- **m1 columns = m2 rows**

Unlike element‑wise multiplication, the two tensors **do not** need to have the same shape.  
They only need to satisfy the column/row rule.

#### Why this rule exists
Matrix multiplication combines rows from the first matrix with columns from the second matrix.  
Each output value is a dot product between:

- one row from matrix 1  
- one column from matrix 2  

If the lengths don’t match, the dot product cannot be computed.

```python
# Notice that the shapes of the tensors differ 
# But m1 columns matches m2 rows 
m1 = torch.tensor([[1,2,3], [4,5,6]])
m2 = torch.tensor([[7,8], [9,10], [11,12]])

# Results will be a 2x2 tensor 
matrix_product = m1@m2 

print(matrix_product)
```

#### Why this is **very** important to understand

Every neural network layer performs
- Output = input @ weights + bias (y = X@W + b)

You will **always** use the '@' operator for this.
